```python
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.optim as optim

from CosineAnnealingWithRestartsLR import CosineAnnealingWithRestartsLR
```

```python
dummy_net_1 = nn.Sequential(nn.Linear(3,2))

optimizer = optim.SGD(dummy_net_1.parameters(),lr=1e-3,momentum=0.9)
scheduler = CosineAnnealingWithRestartsLR(optimizer,T_max=30, eta_min=1e-6)

lrs= []
for i in range(180):
    optimizer.step()
    scheduler.step()
    for param_group in optimizer.param_groups:
        lrs.append(param_group['lr'])
plt.plot(lrs)
plt.show()
```
